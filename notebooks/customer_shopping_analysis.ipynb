we see that Review Rating has 37 missing values

```python
missing_value_percentage = (
    pd.to_numeric(df['Review Rating'], errors='coerce')
    .isna()
    .mean() * 100
)
print(missing_value_percentage)
```

0.94% missing values => median imputation will suffice

```python
df[['Category','Review Rating']]
```

```python
# Fill missing 'Review Rating' with median rating per 'Category'
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Standardize column names
df.columns = df.columns.str.replace(' ', '_').str.lower()
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

### Feature Engineering

```python
# Create age group feature
labels = {'Young Adult', 'Adult', 'Middle Aged', 'Senior'}
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df
```

```python
# Create Frequency purchase days feature
frequency_mapping = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df['purchase_frequency_days']
```

```python
# We want to check if discount applied automatically means promo_code_used (if they are the same)
df[['discount_applied', 'promo_code_used']].head()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

=> we can drop one of the two columns (promo_code_used)

```python
df = df.drop(columns=['promo_code_used'], axis=1)
```

```python
import os
from dotenv import load_dotenv
import pandas as pd
from sqlalchemy import create_engine

load_dotenv()  # looks for .env in project root

PGUSER = os.getenv("PGUSER")
PGPASSWORD = os.getenv("PGPASSWORD")
PGHOST = os.getenv("PGHOST")
PGPORT = os.getenv("PGPORT")
PGDATABASE = os.getenv("PGDATABASE")

engine = create_engine(
    f"postgresql+psycopg2://{PGUSER}:{PGPASSWORD}@{PGHOST}:{PGPORT}/{PGDATABASE}"
)
```

```python
# Load DataFrame to PostgreSQL
df.to_sql(
    name="customer",
    con=engine,
    schema="public",
    if_exists="replace",  # "replace" for first load, "append" for later loads
    index=False,
    method="multi",        # batch insert for speed
    chunksize=10_000       # adjust for large CSVs
)
```

```python
output_path = PROJECT_ROOT / "data" / "processed" / "customer_shopping_behavior_processed.csv"
df.to_csv(output_path, index=False)
```
